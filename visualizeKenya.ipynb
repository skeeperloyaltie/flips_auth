```python
# Kenya Area Visualization using Matplotlib

# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Define the coordinates for Kenya
kenya_area_coordinates = [
    (34.816, 5.000),   # NW point
    (41.000, 5.000),   # NE point
    (41.000, -4.000),  # SE point
    (34.816, -4.000),  # SW point
    (34.816, 5.000)    # Closing the polygon
]

# Unzip the coordinates into two lists
lon, lat = zip(*kenya_area_coordinates)

# Create a plot
plt.figure(figsize=(10, 8))
plt.fill(lon, lat, color='lightblue', alpha=0.5)
plt.plot(lon, lat, marker='o', color='blue', markersize=5)

# Adding titles and labels
plt.title('Geographical Area of Kenya')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show grid
plt.grid()

# Show the plot
plt.show()
```

```python
import folium
from django.contrib.gis.geos import GEOSGeometry

# Define the coordinates for Kenya's borders
kenya_area_coordinates = [
    (5.000, 34.816),   # NW point
    (5.000, 41.000),   # NE point
    (-4.000, 41.000),  # SE point
    (-4.000, 34.816),  # SW point
    (5.000, 34.816)    # Closing the polygon
]

# Create a folium map centered around Kenya
m = folium.Map(location=[-1.2921, 36.8219], zoom_start=6)

# Create a polygon for the entire area of Kenya and add it to the map
folium.Polygon(
    locations=kenya_area_coordinates,
    color='green',
    fill=True,
    fill_color='lightgreen',
    fill_opacity=0.5,
).add_to(m)

# Define the rig data (sensor ID, location, latitude, longitude)
rigs_data = [
    {
        "sensor_id": "Tana Upper",
        "location": "Kone",
        "latitude": -0.07842642962340313,
        "longitude": 38.95772447774984,
    },
    {
        "sensor_id": "Tana Center",
        "location": "Sankuri",
        "latitude": -0.2961679840764966,
        "longitude": 39.53842118297724,
    },
    {
        "sensor_id": "Tana Lower",
        "location": "Sombo",
        "latitude": -0.5678196253853346,
        "longitude": 39.68674985306361,
    },
]

# Add the rig locations as markers on the map
for rig in rigs_data:
    folium.Marker(
        location=(rig['latitude'], rig['longitude']),
        popup=f"{rig['sensor_id']} - {rig['location']}",
        icon=folium.Icon(color='blue')
    ).add_to(m)

# Save the map to an HTML file
m.save('kenya_rigs_map.html')

# Display the map (in a Jupyter notebook, you can display m)
m
```
